## 1. INTRODUCTION 
<br>The FIFA Men’s World Cup, so called the World Cup, is the largest and most prestigious international football tournament, attracting millions of viewers from around the world. With the increasing popularity of football, the need for an efficient and comprehensive database to store and manage the vast amount of data associated with the tournament has become essential.</br>
<br>The aim of this final project is to design a basic database for the World Cup that will store and manage data related to the teams, players, matches, and tournament statistics. The database will provide a platform for users to access information about the tournament, including schedules, results, team standings, and player statistics.</br>
<br> The database design project will apply theoretical knowledge from the SEIS 630-03 Database Management Systems and Design course into practical by carrying out steps of database design such as using normalization, denormalization, building a relational model, ER diagram, transforming data model to database design, then implementing the database and analyzing it using SQL.</br>
<br>The project will cover these following topics:
 - **Analysis of the data requirements**: Chapter 2 begins with an analysis of the data requirements for the FIFA World Cup database. Chapter 3 will involve an examination of the data sources, data elements, and data relationships that are required to support the database.
 - **Normalization/Denormalization**: In Chapter 4, a process of normalization and denormalization will be carried out to organize the data efficiently.
 - **Relational Model**: Chapter 4 also includes the topic of relational model and ER diagram which are developed based on the normalized data.
 - **Database Design**: Chapter 5 will describe transformation of the data model into a database design, including the creation of the database schema, description of database relationships, table dictionary and DDL creating tables.
 - **Database implementation**: Chapter 6 will also involve the implementation of the database and the evaluation of its performance and functionality. It will describe the SQL DDL and DML statements for inserting, updating, and deleting tables.
 - **SQL Views and SQL/PSM**: In the second part of Chapter 6, SQL Views and SQL/PSM will be used for the analysis of the database performance metrics.

## 2. BUSINESS REQUIREMENTS 
The World Cup Database requires information on the 9 main topics: Player, Manager, Country, Confederation, Tournament, Referee, Stadium, Match, and Award.

### 2.1. Requirements for Player 
 
The requirements for Player in the World Cup database are:
-	Each player should have a unique identifier. 
-	Each player should have a name, date of birth.
-	Each player should be associated with a country. However, a player can play two different countries in two tournaments (e.g., Luis Monti played for Argentina in the World Cup 1930, and then played for Italy in the World Cup 1934).
-	Each player should have a position (e.g., goalkeeper, defender, midfielder, forward).
-	Each player should be associated with a tournament, indicating the tournaments they have participated in.
-	Each player may have won individual awards at a tournament, which should be recorded in the awards dataset.

### 2.2. Requirements for Manager
The requirements for Manager in the World Cup database are:
-	Each manager should have a unique identifier.
-	Managers should have a name, date of birth.
-	Each manager should be associated with a team and a country.
-	It should be possible to query the database to retrieve information about all managers associated with a particular team or country.
-	Each manager may be associated with multiple teams and tournaments.

### 2.3. Requirements for Country
The requirements for Country in the World Cup database are as follows:
-	The database should contain a list of countries that are members of the FIFA. 
-	A country may participate in many tournaments.
-	Each country should have a unique identifier.
-	The database should store the home country of each player, manager, and referee.
-	The database should store the country of origin for each stadium where matches are played.

### 2.4. Requirements for Tournament
The requirements for Tournaments in the World Cup database are:
-	Unique ID: each tournament should have a unique identifier in the database to ensure that it can be easily referenced by other entities in the database.
-	Host country: the database should store the name of the country hosting the tournament. A tournament can be hosted by joining host countries. For example, the World Cup 2002 took place in Japan and Korea. A country can organize different tournaments.
-	Winner: the database should store the name of the winning team for each tournament.
-	Dates: the database should store the start and end dates of each tournament.

### 2.5. Requirements for Confederation
The requirements for Confederation in the World Cup database are:
-	Store information about the six FIFA confederations: AFC, CAF, CONCACAF, CONMEBOL, OFC, and UEFA.
-	Each confederation should have a unique ID number.
-	Store the name and code of each confederation.
-	Store the countries that belong to each confederation.
-	Store the referees and teams that belong to each confederation.
-	Allow for querying and reporting on information related to each confederation, including teams, players, referees, and tournaments.

### 2.6. Requirements for Stadium
The requirements for Stadium in the World Cup database are:
-	Each stadium must have a unique identifier.
-	Each stadium must have a name, location (city and country), and capacity.
-	Each stadium must have a list of matches that have been played in it during the World Cup, including the date and time of each match and the teams that played.
-	It should be possible to easily retrieve a list of all stadiums that have hosted matches during a particular tournament or in a particular.

### 2.7. Requirements for Match
The requirements for Match entity in the World Cup database are:
-	Each match must have a unique identifier.
-	Each match must be associated with a tournament, including the tournament stage (group stage, knockout stage, etc.).
-	Each match must have a date and time, location (country, city, and stadium), and a referee assigned to it.
-	Each match involves two teams, and the teams must be identified in the database.
-	A match should have a set of in-match events associated with it, such as goals, bookings, substitutions, and penalty kicks.
-	Each player's appearance in the match must be recorded, including the minute of their appearance and whether they were a starter or substitute.

### 2.8. Requirements for Referee 
The requirements for Referee in the World Cup database are:
-	A unique ID number for each referee.
-	The name and birthday of the referee.
-	The home country of the referee.
-	The confederation of the referee.
-	The list of matches that the referee has officiated in, with details on the teams involved, the stage of the tournament, and the result.

### 2.9. Requirements for Award 
The requirements for Award in the World Cup database are:
-	Store information about each individual award handed out at each tournament.
-	Each award should have a unique ID number.
-	Information about the player who received the award, including their name, team, position, and country.
-	Information about the award itself, including the name of the award, the year it was given, and any other relevant details.
-	The ability to map awards to specific tournaments and matches, including the stage of the tournament in which the award was given.

## 3. DATA COLLECTION AND ANALYSIS <a class="anchor" id="chapter3"></a>

### 3.1. Overview of the database<a class="anchor" id="section_3_1"></a>
The 27 datasets in the database (collected from __[jfjelstul/worldcup: A Comprehensive Database on the FIFA World Cup (github.com)](https://github.com/jfjelstul/worldcup)__ are organized into 5 groups:
-	The first group contains 9 datasets, which provide information about tournaments, FIFA confederations, teams, players, managers, referees, stadiums, matches, and awards. Each unit of observation in this group has a unique ID number.
-	The second group consists of 4 datasets that map teams, players, managers, and referees to tournaments. These datasets include information about qualified teams, squads, manager appointments, and referee appointments.
-	The third group includes 4 datasets that map teams, players, managers, and referees to individual matches. These datasets include information about team appearances, player appearances, manager appearances, and referee appearances.
-	The fourth group contains 4 datasets covering in-match events such as goals, penalty kicks, bookings, and substitutions. Each of these datasets includes the minute of the event and the player(s) and team involved.
-	The fifth and final group consists of 6 datasets covering tournament-level attributes. These datasets include information about host countries, tournament stages, groups, group standings, tournament standings, and individual player awards.

### 5.3. Column Properties for the World Cup Database Design Tables

```python
from sqlalchemy import create_engine, inspect
from prettytable import PrettyTable

inspector = inspect(engine)
def Table_Dictionary(table_name):
    columns = inspector.get_columns(table_name)
    pk_cols = inspector.get_pk_constraint(table_name)['constrained_columns']
    fk_cols = [fk['constrained_columns'][0] for fk in inspector.get_foreign_keys(table_name)]

    table = PrettyTable(['Column Name', 'Type', 'Key', 'NULL Status', 'Remarks'])
    table.align = 'l'

    for column in columns:
        col_name = column['name']
        col_type = column['type']
        is_nullable = column['nullable']
        is_primary_key = col_name in pk_cols
        is_foreign_key = col_name in fk_cols
        key_suffix = ''
        if is_primary_key:
            key_suffix = 'PK'
        elif is_foreign_key:
            key_suffix = 'FK'
        remarks = ''

        table.add_row([col_name, str(col_type), key_suffix, 'NOT NULL' if not is_nullable else '', remarks])
    print(f"Table {table_name.upper()}")
    print(table)
    print('\n')
    
Table_Dictionary('confederation')
Table_Dictionary('country')
Table_Dictionary('host_country')
Table_Dictionary('country_appearance')
Table_Dictionary('tournament')
Table_Dictionary('qualified_team')
Table_Dictionary('squad')
Table_Dictionary('player')
Table_Dictionary('player_appearance')
Table_Dictionary('manager')
Table_Dictionary('manager_appointment')
Table_Dictionary('referee')
Table_Dictionary('referee_appointment')
Table_Dictionary('stadium')
Table_Dictionary('match')
Table_Dictionary('goal')
Table_Dictionary('booking')
Table_Dictionary('award')
Table_Dictionary('award_winner')
```

#### 6.2.2. Create VIEW

```python
%%sql 
CREATE VIEW HOST_COUNTRY_INFO AS
SELECT 
    H.tournament_id, 
    host_country_id, 
    country_name,
    (CASE 
        WHEN country_name=winner THEN 'champion'
        ELSE performance
    END) AS
    performance, 
    winner,
    (CASE 
        WHEN country_name=winner THEN 'Yes'
        ELSE 'No'
    END) AS host_won
FROM host_country H
JOIN COUNTRY C
    ON H.host_country_id = C.country_id
JOIN qualified_team Q
    ON H.tournament_id = Q.tournament_id AND H.host_country_id = Q.country_id
JOIN TOURNAMENT T
    ON T.tournament_id = H.tournament_id
ORDER BY tournament_id
```

```python
%sql select * from HOST_COUNTRY_INFO
```

#### 6.2.3. Create User-defined function

```python
%%sql
CREATE OR REPLACE FUNCTION getfullname (
    given_name IN VARCHAR2,
    family_name IN VARCHAR2
)
RETURN VARCHAR2
AS
BEGIN
    RETURN (CASE 
                WHEN given_name = 'not applicable' THEN family_name
                ELSE given_name || ' ' || family_name
            END);
END;
```

#### 6.2.4. Examples of SQL queries
##### Find top 10 players who played most games in the World Cup

```python
%%sql
CREATE VIEW MOST_PLAYED AS
SELECT DISTINCT A.player_id, 
        getfullname(given_name, family_name) as full_name, 
        C.country_name, COUNT(DISTINCT A.match_id) as number_of_matches
FROM PLAYER_APPEARANCE A
JOIN PLAYER P
    ON A.player_id = P.player_id
JOIN SQUAD S
    ON S.player_id = P.player_id
JOIN COUNTRY C
    ON C.country_id = S.country_id
GROUP BY A.player_id, getfullname(given_name, family_name), C.country_name
ORDER BY count(distinct match_id) DESC
```

```python
%sql SELECT * FROM MOST_PLAYED WHERE ROWNUM <= 10
```

There is a duplicate at player_id = P-09502, so we need to modify the query to get uniquely players by adding all countries into a new column country_list.

```python
%%sql 
SELECT  player_id, 
        full_name, 
        LISTAGG(DISTINCT country_name, '/') WITHIN GROUP (ORDER BY country_name) AS country_list,
        number_of_matches
FROM MOST_PLAYED
GROUP BY player_id, full_name, number_of_matches
ORDER BY number_of_matches DESC
FETCH FIRST 10 ROWS ONLY
```

Now, we can create Data Visualization with it

```python
#Use SQL query to get data from database
sql_query = '''
    SELECT player_id, 
           full_name, 
           number_of_matches, 
           LISTAGG(DISTINCT country_name, ', ') WITHIN GROUP (ORDER BY country_name) AS country_list
    FROM MOST_PLAYED
    GROUP BY player_id, full_name, number_of_matches
    ORDER BY number_of_matches DESC
    FETCH FIRST 10 ROWS ONLY
'''

# Execute the query and get the results in a pandas dataframe
df = pd.read_sql_query(sql_query, engine)
df['country_list'] = df['country_list'].replace('Germany, West Germany', 'Germany')


# Draw graph
sns.set(rc={'figure.figsize':(8, 4)})
sns.barplot(x='number_of_matches', y='full_name', data=df, hue='country_list', dodge=False)
plt.title('Top 10 Most Played Players in the World Cup')
plt.xlabel('Number of Matches')
plt.ylabel('Player')
plt.legend(title="Country List", fontsize='xx-small') # adjust legend font size
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

##### Find players who played for two different countries

```python
%%sql 
SELECT
    getfullname(P. given_name, P.family_name) as full_name, COUNT(DISTINCT C.country_name) as number_of_countries,
    LISTAGG(DISTINCT C.country_name, ', ') WITHIN GROUP (ORDER BY C.country_name) as country_list
FROM SQUAD S
JOIN PLAYER P
    ON S.player_id = P.player_id
JOIN COUNTRY C
    ON S.country_id = C.country_id
GROUP BY getfullname(P. given_name, P.family_name)
HAVING COUNT(DISTINCT C.country_name) > 1
ORDER BY COUNT(DISTINCT C.country_name) DESC
```

##### Goals scored by Mexico, United States and Canada in the World Cup

```python
#Query the data
sql_query = '''
    SELECT goal_id, minute_regulation, minute_stoppage, penalty, 
    match_id, G.player_id, country_name, G.country_id AS for_team_id, S.country_id as from_team_id,
    (CASE 
        WHEN G.country_id = S.country_id AND penalty = 0 THEN 'Goal'
        WHEN G.country_id = S.country_id AND penalty = 1 THEN 'Penalty'
        ELSE 'Own Goal'
    END) AS goal_type
    FROM GOAL G
    JOIN COUNTRY C 
        ON G.country_id = C.country_id
    JOIN SQUAD S
        ON S.player_id = G.player_id
    WHERE country_name in ('Mexico', 'United States', 'Canada')
    
'''

# Execute the query and get the results in a pandas dataframe
dt = pd.read_sql_query(sql_query, engine)
dt.head()
```

```python
#Data visualization
sns.set(style="white", rc={'figure.figsize':(6, 4)})

# Create the swarm plot and get the Axes object
ax = sns.swarmplot(y='minute_regulation', data=dt, x='country_name', hue='goal_type')

# Add the horizontal lines
ax.axhline(45, color='lightgrey', linewidth=1, xmin=0.05, xmax=0.95)
ax.axhline(90, color='lightgrey', linewidth=1, xmin=0.05, xmax=0.95)

# Set the plot title and labels
ax.set_title('Goals scored by Mexico, United States and Canada in all World Cup matches')
ax.set_ylabel('Minute of the match')
ax.yaxis.set_ticks(range(0, 121, 15))

# Remove the x label "country_name"
ax.set_xlabel('')

# Remove the frame line of the graph
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['right'].set_visible(False)

# Move the legend to the top and outside of the graph
ax.legend(fontsize='xx-small', bbox_to_anchor=(1.05, 0.6), loc=2, borderaxespad=0.)
# Show the plot
plt.show()
```
